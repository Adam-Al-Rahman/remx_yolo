```python
import cv2
import numpy as np
```

```python
# load the video frames
frame1 = cv2.imread("frame1.jpg")
frame2 = cv2.imread("frame2.jpg")
```

```python
bbox1 = (256, 390, 950 - 256, 902 - 390)
print(bbox1)
```

```python
# for TrackerCSRT_create() install opencv-contrib-python module
tracker = cv2.TrackerCSRT_create()
tracker.init(frame1, bbox1)
```

```python
# Track the object of interest in the second frame
success, bbox2 = tracker.update(frame2)
bbox2 = (115, 374, 742, 886)
print(bbox2)
```

```python
# Calculate the distance between the two positions of the object
x1, y1, x2, y2 = bbox1
cx1, cy1 = (x1 + x2)/2, (y1 + y2)/2

x1, y1, x2, y2 = bbox2
cx2, cy2 = (x1 + x2)/2, (y1 + y2)/2

distance = np.sqrt((cx2 - cx1)**2 + (cy2 - cy1)**2)
distance
```

```python
# Draw the object bounding box on the second frame
bbox2 = tuple(map(int, bbox2))
cv2.rectangle(frame2, bbox2, (0, 255, 0), 2)
print(bbox2)
```
